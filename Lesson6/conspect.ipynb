```python
#График процентра реализованныйх сделок по дням (CR по дням)
```

```python
cr_by_day_rolling = cr_by_day = order_leads.groupby(['Date'])\
    .agg({'Converted': 'mean'})\
    .rename(columns={'Converted': 'CR'})\
    .sort_values('Date')\
    .rolling(30).mean()
```

```python
ax = cr_by_day_rolling.plot()
ax.set_title('Convertion rate by day\n')
ax.set_ylabel('Convertion rate')
ax.set_xlabel('\nDate')
ax.set_yticklabels(['{:.0%}'.format(n) for n in ax.get_yticks()])
sns.despine()
```

```python
def get_plot(df, x_label='', y_label='', title=''):
    ax = df.plot()
    ax.set_title('title\n')
    ax.set_ylabel('y_label')
    ax.set_xlabel('\nx_label')
    ax.set_yticklabels(['{:.0%}'.format(n) for n in ax.get_yticks()])
    sns.despine()
```

```python
get_plot(cr_by_day_rolling, x_label = 'Date', y_label ='Convertion rate', title = 'Convertion rate by day')
```
