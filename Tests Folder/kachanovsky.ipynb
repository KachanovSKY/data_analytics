### Задача ТЗ:

За каждую заключенную сделку менеджер получает бонус, который рассчитывается
следующим образом.

1. За новые сделки менеджер получает 7 % от суммы, при условии, что статус оплаты «ОПЛАЧЕНО», а также имеется оригинал подписанного договора с клиентом (в рассматриваемом месяце).
2. За текущие сделки менеджер получает 5 % от суммы, если она больше 10 тыс., и 3 % от суммы, если меньше. При этом статус оплаты может быть любым, кроме «ПРОСРОЧЕНО», а также необходимо наличие оригинала подписанного договора с клиентом (в рассматриваемом месяце).

Бонусы по сделкам, оригиналы для которых приходят позже рассматриваемого месяца, считаются остатком на следующий период, который выплачивается по мере прихода оригиналов. Вычислите остаток каждого из менеджеров на 01.07.2021.

```python
df
```

Из опиания задания я понял что алгоритм решения выглядит следующим образом:
- Нам надо смотреть на сделки проведенные раньше 01.07.21, у которых оригинал договора пришел позже этой даты. Так дельни, отведенные менеджерам будут считаться отсатком.

```python
df.query('status == "Май 2021" or status == "Июнь 2021" or status == "Июль 2021"')
```

```python
df_remains = df.iloc[1:258]
```

```python
df_remains = df_remains.drop(index=129)
df_remains
```

```python
df_remains.iloc[125:135] # проверил что Лейбл Июля стерся корректно
```

```python
df_remains = df_remains.query("receiving_date > '2021-07-01'")
df_remains.reset_index(drop= True , inplace= True )
df_remains
```

```python
df_remains['sale_bonus'] = np.NaN
for i in range(df_remains.shape[0]):
    if ((df_remains['new/current'][i] == 'новая') & (df_remains['status'][i] == 'ОПЛАЧЕНО') & (df_remains['document'][i] == 'оригинал')):
        df_remains['sale_bonus'][i] = df_remains['sum'][i]* 0.07
    elif ((df_remains['new/current'][i] == 'текущая') & (df_remains['status'][i] != 'ПРОСРОЧЕНО') & (df_remains['document'][i] == 'оригинал')):
        if (df_remains['sum'][i]> 10000):
            df_remains['sale_bonus'][i] = df_remains['sum'][i]* 0.05
        else:
            df_remains['sale_bonus'][i] = df_remains['sum'][i]* 0.03
```

```python
df_remains
```

```python
df_remains.groupby('sale').agg({'sale_bonus':'sum'})
```
