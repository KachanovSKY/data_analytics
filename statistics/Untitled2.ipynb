```python
cars['CarMark'] = cars.CarName.apply(lambda x: x.split(' ')[0].lower())
```

```python
cars.CarMark.nunique()
```

```python
cars.CarMark = cars.CarMark.apply(lambda x: x.replace('maxda', 'mazda'))
cars.CarMark = cars.CarMark.apply(lambda x: x.replace('porcshce', 'porsche'))
cars.CarMark = cars.CarMark.apply(lambda x: x.replace('toyouta', 'toyota'))
cars.CarMark = cars.CarMark.apply(lambda x: x.replace('vokswagen' or 'vw', 'volkswagen'))
```

```python
df_dummy = pd.get_dummies(data=cars[['CarMark', 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginetype', 'cylindernumber']], drop_first = True)
```

```python
cars_2 = cars.drop(['CarMark', 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginetype', 'cylindernumber'], axis=1)
```

```python
cars_3 = cars[['wheelbase', 'carlength', 'carwidth', 'carheight', 'boreratio', 'stroke', 'compressionratio', 'price', 'horsepower']]
```

```python
all_cars = pd.concat([df_dummy, cars_3], axis=1)
```

```python
all_cars.horsepower
```

```python
cars_3
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf 

# способ первый
# Y = одномерный массив с ЗП, X - массив со всеми нужными нам НП

x = sm.add_constant(all_cars.horsepower)  # добавить константу, чтобы был свободный член
model = sm.OLS(all_cars.price, x)  # говорим модели, что у нас ЗП, а что НП
results = model.fit()  # строим регрессионную прямую
print(results.summary())  # смотрим результат
```
